## Education Categorization

## Importing the libararies

```python
import pandas as pd
import numpy as np
```

## Importing the file

```python
data = pd.read_json("Data/new_jobs-7-7-22.json")
data.head()
```

## Extracting the Educations

```python
educations = data.iloc[:, 3:4].values
educations
```

## Education Categories

```python
EDUCATIONS = [
    "bachelor-computer engineering",
    "master-computer engineering",
    "phd-computer engineering",
    
    "others",
]
```

```python
new_educations = []
# print(experiences)
for education in educations:
    new_education = []
    for item in education[0]:
        new_item = {"university": item['university'], "major": ""}
        
        if ("bachelor" in item['major'].lower().split() or "bachelors" in item['major'].lower().split()) and ("computer" in item['major'].lower().split() or "it" in item['major'].lower().split() or ("information" in item['major'].lower().split() and "technology" in item['major'].lower().split())):
            new_item["major"] = EDUCATIONS[0]
        elif ("master" in item['major'].lower().split() or "masters" in item['major'].lower().split()) and "computer" in item['major'].lower().split():
            new_item["major"] = EDUCATIONS[1]
        elif ("phd" in item['major'].lower().split()) and "computer" in item['major'].lower().split():
            new_item["major"] = EDUCATIONS[2]   
        else:
            new_item["major"] = EDUCATIONS[3]
        new_education.append(new_item)
    new_educations.append(np.array(new_education))
```

```python
data["Educations"] = new_educations
```

## Exporting to file

```python
data.to_csv("Data/new_educations-7-7-22_1.csv")
data.to_json("Data/new_educations-7-7-22_1.json")
```
